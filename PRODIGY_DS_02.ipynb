```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv(r'C:\Users\gowth\Downloads\PRODIGY_internship\dataset\task 2\titanic\train.csv') 
```

```python
train_df.info()
```

```python
train_df.head()
```

```python
train_df.isnull().sum()
```

```python
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)  # Fill Age with median
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)  # Fill Embarked with mode
train_df.drop(columns=['Cabin'], inplace=True)  # Drop Cabin due to too many missing values
```

```python
train_df.isnull().sum()
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Countplot for Survival
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=train_df, palette='coolwarm')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

```

```python
# Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Sex', data=train_df, palette='coolwarm')
plt.title("Survival by Gender")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Pclass', data=train_df, palette='coolwarm')
plt.title("Survival by Passenger Class")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

```

```python
# Age distribution by Survival
plt.figure(figsize=(8,5))
sns.histplot(train_df, x='Age', hue='Survived', kde=True, bins=30, palette='coolwarm')
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(8,5))
sns.histplot(train_df['Fare'], bins=30, kde=True, color='purple')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate based on SibSp (Number of Siblings/Spouses aboard)
plt.figure(figsize=(6,4))
sns.countplot(x='SibSp', hue='Survived', data=train_df, palette='coolwarm')
plt.title("Survival by SibSp (Siblings/Spouses Aboard)")
plt.xlabel("SibSp")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate based on Parch (Number of Parents/Children aboard)
plt.figure(figsize=(6,4))
sns.countplot(x='Parch', hue='Survived', data=train_df, palette='coolwarm')
plt.title("Survival by Parch (Parents/Children Aboard)")
plt.xlabel("Parch")
plt.ylabel("Count")
plt.show()
```

# More women survived compared to men
# Higher-class passengers had a better survival rate
# People with family (Parch & SibSp) had different survival rates
# Age played a role in survival
